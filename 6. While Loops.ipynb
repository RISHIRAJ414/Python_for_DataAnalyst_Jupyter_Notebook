```python
#Explanation
#The while loop is used to repeatedly execute a block of code as long as a given condition is True. 

#Here's how it relates to the flowchart:
#Key Steps:
#1.Start: The loop begins execution.
#2.Condition Check: Before each iteration, the loop checks the condition.
# *If the condition evaluates to True, the loop body is executed.
# *If the condition evaluates to False, the loop terminates.
#3.Execute Loop Body: The block of code inside the loop is run when the condition is True.
# *Repeat: After executing the loop body, the condition is checked again.

#Difference Between For loop and While loop : 
# -Condition Check :
# *for loop: Iterates over each item in a sequence until the sequence is exhausted (no explicit condition is required).
# *while loop: Requires a condition that is checked before each iteration. The loop executes as long as the condition is True.
# -Infinite Loop Risk :
# *for loop: Rare, because it iterates over a predefined sequence.
# *while loop: Higher risk if the condition is not updated correctly or is always True.
# -Both loops support break (to exit the loop) and continue (to skip to the next iteration).

#When to Use:
# -Scenario	                                  -Recommended Loop
# 1.Iterating over a sequence or range	        for loop
# 2.Looping until a condition is met	        while loop
# 3.Unknown number of iterations	            while loop
# 4.Known number of iterations		            for loop
```

```python
number = 0
while number < 5:  #Here loop will run until this Condition is True and when the condition gets False , it will break the loop and will exit.
    print(number)
    number = number + 1
```

```python
#Fibonacci Series 
```

```python
# Initialize variables
a, b = 0, 1
limit = 5
print("Fibonacci series up to", limit, ":")

# While loop to generate Fibonacci series
while a <= limit:
    print(a, end=" ")
    a, b = b, a + b
```

```python
#Break statement
#Using break statement we can break or stop the loop even if the condition in True.
```

```python
number = 0
while number < 5:  
    print(number)
    if number == 3:
        break
    number = number + 1
```

```python
#While loop + else statement
```

```python
number = 0
while number < 5:  
    print(number)
    if number == 6:
        break
    number = number + 1
else:
    print('Number no longer < 5')
```

```python
#Using While loop with if else statement.
#Code to print all multiples of 2 less than 30 .
```

```python
number = 0
while number < 30:  
    if number % 2 == 0:
        print(number)
    number += 1  # Increament number to avoid infinite loop
else:
    print('Number is no longer < 30')  # Properly indented

#Explanation:
# *Condition: The loop runs as long as number is less than 30.
# *Check for Multiples of 2: number % 2 == 0 checks if the number is a multiple of 2.
# *Increment: number += 1 ensures the loop progresses.
# *Else Block: Executes after the loop ends, confirming the termination condition.
```

```python
#Continue statement
#If continue statement is triggered then it rejects all the current iterations of the loop and will go to the Next iteration.
#Basically it jumps back to the starting and start running it even if there is current iteration which stops the loop.

#Explanation:
# *The continue statement in Python is used to skip the current iteration of a loop and proceed with the next iteration. 
# *It is commonly used within loops like for or while when you want to bypass the rest of the code inside the loop for a particular iteration and move to the next one.

#How continue Works:
# *When Python encounters the continue statement, it skips the remaining code in the current iteration and moves directly to the next iteration of the loop.
#Example:
    for number in range(1, 6):
    if number == 3:
        continue  # Skip the rest of the code when number is 3
    print(number)
#Output:
#    1
#    2
#    4
#    5

#In this example:
# *When the loop reaches the number 3, the continue statement is executed.
# *The print(number) statement is skipped for that iteration and the loop continues with the next number (i.e., 4).

#Use Cases for continue:
# *Skipping unwanted values: You might want to skip processing for certain values in a loop.
# *Reducing complexity: It can help avoid deep nesting of if statements by skipping iterations early.

#Example - Skipping Even Numbers:
for number in range(1, 10):
    if number % 2 == 0:
        continue  # Skip even numbers
    print(number)
#Output:
#    1
#    3
#    5
#    7
#    9
#In this case, the continue statement ensures that only odd numbers are printed by skipping the even numbers.

#Summary:
# *Purpose: The continue statement is used to skip the current iteration of a loop and proceed with the next iteration.
# *Common Use Cases: Skipping certain conditions (like specific numbers or values) or reducing unnecessary code execution inside loops.

```

```python
number = 0
while number < 5:  
    number = number + 1
    if number == 3:
        continue
    print(number)
else:
    print('Number no longer < 5')
```
